```python
## Alunos: Matheus Petters Bevilaqua e Natália Sens Weise

# imports
from bs4 import BeautifulSoup
import requests
import time
import json

links = [
    "https://www.amazon.com.br/Carregador-Pilhas-com-pilhas-Duracell/product-reviews/B07X4JG2G4/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews",
    "https://www.amazon.com.br/Conjunto-Brinox-7001-101-Vermelho/product-reviews/B01MY00D73/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews",
    "https://www.amazon.com.br/Echo-Pop-Cor-Preta/product-reviews/B09WXVH7WK/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews",
    "https://www.amazon.com.br/Fone-Ouvido-Vibe-Preto-PH322/product-reviews/B086MLR6JS/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews",
    "https://www.amazon.com.br/Novo-Kindle-lan%C3%A7amento-2022-armazenamento/product-reviews/B09SWV1FSS/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews",
    "https://www.amazon.com.br/Conjunto-Santoku-Oster-Baldwyn-inoxid%C3%A1vel/product-reviews/B07176S6CR/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews",
    "https://www.amazon.com.br/Pen-Drive-Titan-Multilaser-Acess%C3%B3rios/product-reviews/B07D2GPZ3M/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews",
    "https://www.amazon.com.br/Kit-com-Camisetas-Masculina-Part-B/product-reviews/B091PDL2MH/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews",
    "https://www.amazon.com.br/Mochila-Executiva-Vintage-Notebook-Multifuncional/product-reviews/B0C7SFPRR6/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews",
    "https://www.amazon.com.br/Computador-Completo-N%C3%BAcleos-Strong-Tech/product-reviews/B0BDW2W9S5/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews",
    ]

output = []

## criando um laço de repetição pra pegar comentarios de cada produto da lista
for i in range(len(links)):
  ## objeto que vai salvar os comentarios de cada produto
  product = {
    "product": "produto"+str(i+1),
    "comments": []
    }

  html = requests.get(links[i]).content
  print(links[i])
  # instanciando a soup pra ler o html
  soup = BeautifulSoup(html, 'lxml')
  comm_table = soup.find_all("span", class_="a-size-base review-text review-text-content")
  # print(comm_table)

  ## laço pra pegar os 10 comentarios visiveis na pg, e colocar na lista do obj json desse produto
  for k in range (10):
    ## aqui pega uma substring pq ficava aparecendo os "\n" no começo e no fim
    comment = {"text": comm_table[k].text[1:-1]}
    if comm_table[k].text[1:-1].find("\n\n\n\n                    A mídia não pôde ser carregada.\n                \n\n\n\n\xa0")  == 0:
      ## cortando o começo de comentarios que tenham midia inclusisa
      comment = {"text": comm_table[k].text[78:-1]}
    product["comments"].append(comment)
  ## sleep pra n "sobrecarregar" o site ou questoes legais talvez
  time.sleep(2)
  print(product)
  output.append(product)

jsonString = json.dumps(output)
jsonFile = open("data.json", "w")
jsonFile.write(jsonString)
jsonFile.close()
```
