```python
## Matheus Petters, Nat√°lia Sens Weise
from nltk.stem.snowball import SnowballStemmer
import nltk
import json

nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('portuguese')

f = open('data.json')
data = json.load(f)

## AQUI > REMOVER PONTUA√á√ÉO E EMOJES POR REGEX

comments = []
for i in data:
  for comment in i["comments"]:
    comments.append(comment["text"].split())

## removendo pontua√ß√£o e emojis

## removendo stopwords
for comment in comments:
  for word in stopwords:
    if word in comment:
      comment.remove(word)
# print(comments)

## stems
stemmer = SnowballStemmer("portuguese")

for comment in comments:
  for word in comment:
    print(stemmer.stem(word))


f.close()
```
